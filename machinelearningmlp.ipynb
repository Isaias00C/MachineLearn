```python
# importar o dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

headernames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

dataset = pd.read_csv(path, names = headernames)
dataset.head()

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
# a parte de escalomento é importante sempre, pois valores muito discrepantes podem interferir no resultado final da ia
```

```python
# modelo de ia
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(max_iter=3000)

# model.fit(X_train, y_train)

from sklearn.model_selection import GridSearchCV
```

```python


param_grid = {
    "hidden_layer_sizes": [x for x in range(2,16)],
    'activation': ['relu'],
    'solver': ['adam'],
    'alpha': [0.0001],
    'learning_rate': ['constant', 'adaptive']
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Melhores parâmetros:", grid_search.best_params_)
```

```python
# resultados
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)

result = confusion_matrix(y_test, y_pred) # uma matrix para mostrar os casos acertados, no caso a diagonal principal sao os casos acertados

from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=result, display_labels=set(y))
disp.plot(cmap='Blues')
plt.title('Matriz de Confusão')
plt.show()

result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)

result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)

result3 = f1_score(y_test, y_pred, average=None)
print("F1-score:", result3)
```
